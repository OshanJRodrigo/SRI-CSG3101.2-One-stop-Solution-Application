```python
# Define the number of lines to read
#n_lines = 100

# Read the specified number of lines into a list of dictionaries
#with open(JSON_CUSTOMER_DATA_FILE, "r") as file:
#    data = [json.loads(next(file)) for _ in range(n_lines)]

# Read the JSON lines file directly into a pandas DataFrame
#df = pd.read_json(JSON_DATA_FILE, lines=True)

# Convert the list of dictionaries into a DataFrame
#customers_df = pd.DataFrame(data)

# Display the first few rows
#print(customers_df.head())


def get_data(filename, num_lines=100):
    # Initialize an empty list to store selected attributes
    filtered_data = []

    # Open and process JSON file line by line
    with open(filename, 'r') as file:
        for i, line in enumerate(file):
            if i == num_lines:  # Stop after reading the specified number of lines
              break
            record = json.loads(line)
            # Extract only specific attributes
            filtered_data.append({'customerId': record['user_id'], 'employeeId': record['business_id']})


    # Create a DataFrame
    return pd.DataFrame(filtered_data)
```

```python
# Initialize Firestore client
db = firestore.Client()
```

```python
# Function to get 'chargePerHour' attribute from Firestore for each employeeId
def get_chargePerHour_from_firestore(employeeId, db):
    # Reference to the document with the given employeeId
    doc_ref = db.collection('employees').document(employeeId)

    # Fetch the document
    doc = doc_ref.get()

    if doc.exists:
        # Get the document data
        data = doc.to_dict()
        # Return the 'chargePerHour' attribute (or None if it doesn't exist)
        return data.get('chargePerHour', 2000) #None)
    else:
        return 2000 #None  # If the document doesn't exist
```

```python
# Add 'chargePerHour' column to the DataFrame
bookings_df['chargePerHour'] = bookings_df['employeeId'].apply(lambda x: get_chargePerHour_from_firestore(x, db))
```
