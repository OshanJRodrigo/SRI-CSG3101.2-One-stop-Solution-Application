```python
JSON_FILE = '/content/drive/My Drive/yelp_academic_dataset_review.json'
```

```python
def get_data(filename, num_lines=100):
    # Initialize an empty list to store selected attributes
    filtered_data = []

    # Open and process JSON file line by line
    with open(filename, 'r') as file:
        for i, line in enumerate(file):
            if i == num_lines:  # Stop after reading the specified number of lines
              break
            record = json.loads(line)
            # Extract only specific attributes
            filtered_data.append({'id': record['review_id'], 'customerId': record['user_id'], 'employeeId': record['business_id'], 'stars': record['stars']})


    # Create a DataFrame
    return pd.DataFrame(filtered_data)
```

```python
df = get_data(JSON_FILE)
```

```python
# Convert DataFrame to JSON
processed_data = df.to_dict(orient='records')
```

```python
print(processed_data[0])
```

```python
# Initialize Firestore client
db = firestore.Client()
```

```python
# Firestore collection name
collection_name = "reviews"
```

```python
# Upload processed data to Firestore
for doc in processed_data:

    # Extract the 'id' before removing it from the document body
    doc_id = doc.pop('id', None)

    # Upload the document without the 'id' in the body
    doc_ref = db.collection(collection_name).document(doc_id)
    doc_ref.set(doc)
```
