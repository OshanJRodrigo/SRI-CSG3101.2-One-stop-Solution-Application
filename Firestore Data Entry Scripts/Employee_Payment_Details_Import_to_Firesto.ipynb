```python
JSON_DATA_FILE = '/content/drive/My Drive/yelp_academic_dataset_business.json'
```

```python
# Define the number of lines to read
n_lines = 100

# Read the specified number of lines into a list of dictionaries
with open(JSON_DATA_FILE, "r") as file:
    data = [json.loads(next(file)) for _ in range(n_lines)]

# Read the JSON lines file directly into a pandas DataFrame
#df = pd.read_json(JSON_DATA_FILE, lines=True)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Display the first few rows
print(df.head())
```

```python
print(len(df)) # total number of entries
```

```python
# Filter rows where 'categories' is not null
df = df[df['categories'].notnull()]

# Select specific columns
df = df[['business_id', 'name']]

# Display the result
print(df.head())
```

```python
print(len(df))  # number of entries after removing 103 rows where 'categories' have null value
```

```python
# Rename columns
df = df.rename(columns={'business_id': 'employeeId', 'name': 'accountHolderName'})

# Display the result
print(df.head())
```

```python
# Function to generate random bank account numbers
def generate_bank_account_number(length=10):
    # Generates a random bank account number of the given length
    return ''.join(np.random.choice([str(i) for i in range(10)], size=length))
```

```python
df['accountNumber'] = [generate_bank_account_number() for _ in range(len(df))]

# Display the result
print(df.head())
```

```python
# List of example bank names
bank_names = [
    'First National Bank', 'Global Trust Bank', 'Bank of the Future',
    'Citizens Bank', 'Sunshine Financial', 'Pioneer Savings Bank',
    'Union Credit Union', 'Capital City Bank', 'Blue Ridge Bank',
    'Harborview Bank', 'Summit Bank', 'National Savings Bank'
]

# Function to generate a random bank name from the list
def generate_bank_name():
    return np.random.choice(bank_names)
```

```python
# Generate random bank names for each row
df['bankName'] = [generate_bank_name() for _ in range(len(df))]

# Display the result
print(df.head())
```

```python
# Function to generate a random bank code (e.g., 6-digit code)
def generate_bank_code():
    return np.random.randint(100000, 999999)
```

```python
df['bankCode'] = [generate_bank_code() for _ in range(len(df))]

# Display the result
print(df.head())
```

```python
# Convert DataFrame to JSON
processed_data = df.to_dict(orient='records')
```

```python
# Initialize Firestore client
db = firestore.Client()
```

```python
# Firestore collection name
collection_name = "employeeBankAccountDetails"
```

```python
# Upload processed data to Firestore
for doc in processed_data:

  # Reference to your Firestore collection
  collection_ref = db.collection(collection_name)

  # Add the document with an auto-generated ID
  doc_ref = collection_ref.add(doc)
```
